```python
import requests
from bs4 import BeautifulSoup as soup
```

```python
header = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.86 Safari/537.36",
         'Referer':'https://www.zillow.com/manhattan-new-york-ny/rentals/?searchQueryState=%7B%22pagination'}
```

```python
url='https://www.zillow.com/manhattan-new-york-ny/rentals'
```

```python
html= requests.get(url=url,headers=header)
html.status_code
```

```python
bsobj=soup(html.content,'lxml')
bsobj
```

```python
price_list=[]
for price in bsobj.findAll('div', class_= "PropertyCardWrapper__StyledPriceGridContainer-srp-8-100-8__sc-16e8gqd-0 gZmMEQ"):
    price_list.append(price.text)
price_list
```

```python
address = []
for adr in bsobj.findAll('a',class_='StyledPropertyCardDataArea-c11n-8-100-8__sc-10i1r6-0 dtcWyn property-card-link'):
    address.append(adr.text)
address
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(price_list,columns=['prices'])
df['Address']=address
```
